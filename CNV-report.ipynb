```python
import urllib.request
import pandas
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
from sklearn.decomposition import PCA
import os
import operator
```

```python
data = pandas.read_csv('CNV/results.txt', header=0, names=['start', 'end'])
```

```python
data = data[data['start'] > 1495300100]
```

```python
data['responseTime'] = data['end'] - data['start']
```

```python
plt.scatter(data['start'], data['responseTime'], s=1)
plt.xlabel('Timestamp of request (s)')
plt.ylabel('Response time (s)')
plt.show()
```

```python
smoothInterval = 10
y,binEdges=np.histogram(data['end'],bins=int((data['end'].max() - data['end'].min())/smoothInterval))
bincenters = 0.5*(binEdges[1:]+binEdges[:-1])
plt.plot(bincenters,y/smoothInterval,'-')
plt.xlabel('Timestamp of request (s)')
plt.ylabel('Throughput (Requests answered/second)')
plt.show()
```
